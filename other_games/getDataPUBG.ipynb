```python
from pubg_key import key
import requests
```

```python
session = requests.Session()
session.headers.update({
            "Authorization": "Bearer "+key,
            "Accept": "application/vnd.api+json"
        })

player = 'Brungilda'
platform = "steam"
response = session.get("https://api.pubg.com/shards/"+platform+"/players?filter[playerNames]="+player)
response.json()
```

```python
id = response.json()['data'][0]['id']
id
```

```python
matches = response.json()['data'][0]['relationships']['matches']['data']
matches
```

```python
match = matches[0]['id']
session.get("https://api.pubg.com/shards/steam/matches/"+match).json()
```
